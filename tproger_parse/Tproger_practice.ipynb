```python
from selenium import webdriver
# from selenium.webdriver.support.ui import WebDriverWait
# from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from time import sleep
driver_path = "c:/Users/admin/chromedriver"


def run(driver_path):

    parse_dict = {}  # возвращаемый словарь с вопросами в качестве ключа и списками ответов в качестве значений
    right_answers = []  # список для сохранения правильных ответов

    def wait_click(class_name):
        while True:
            try:
                button = driver.find_element_by_class_name(class_name)
                button.click()  # Doesn't work on slow connections as it must wait untill it's clickable
                break
            except:
                sleep(1)

    driver = webdriver.Chrome(driver_path)
    driver.get('https://tproger.ru/quiz/real-programmer/')
    wait_click("quiz_button")

    for n in range(1, 13):
        question = driver.find_element_by_xpath(
            f'//*[@id="quiz_item_{254+n}"]/div[1]').text
        print(f'Вопрос №{n}')
        print(question)
        answer = driver.find_element_by_xpath(
            f'//*[@id="quiz_item_{254+n}"]/div[2]/div[1]')
        answer.click()
        sleep(2)
        answers = driver.find_elements_by_class_name("answer")
        answer_list = []
        for i in answers:
            if i.text:
                answer_list.append(i.text)
                print(i.text)
        parse_dict[question] = answer_list
        right_ans = driver.find_elements_by_class_name(
            "answer.correct_answer")[n-1].text
        print('correct answer:', right_ans)
        right_answers.append(right_ans)
        print(n, 'done')
        next_button = driver.find_element_by_xpath(
            f'//*[@id="quiz_item_{254+n}"]/button')
        next_button.click()

    driver.close()
    driver.quit()

    return parse_dict, right_answers


[parse_dict, right_answers] = run(driver_path)
```
